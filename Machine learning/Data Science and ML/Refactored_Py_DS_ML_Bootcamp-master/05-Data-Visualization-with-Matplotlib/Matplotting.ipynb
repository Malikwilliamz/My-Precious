```python
import matplotlib.pyplot as plt
```

```python
#You can go onto the matplotlib website and get code for making different types of graphs
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y  = x**2
```

```python
#Functional Method of creating a MatplotLib Plot
plt.plot(x,y)
plt.title("FIRST PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```
